# 3. Topic modelling

1. Latent Dirichlet Allocation
2. Non-negative Matrix Factorization

```python
def print_top_words(model, feature_names, n_top_words):
  for index, topic in enumerate(model.components_):
    message = "\nTopic #{}:".format(index)
    message = " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 : -1]])
    print(message)
    print("="*70)
```

### 3a. Preprocessing steps 합치기

```python
lemm = WordNetLemmatizer()

# CountVectorizer class에 lemmatizer 합치기(extend the class)
class LemmaCountVectorizer(CountVectorizer):
  def build_analyzer(self): # 기존 CountVectorizer 내에 있던 build_analyzer를 재정의
    analyzer = super(LemmaCountVectorizer, self).build_analyzer() # 부모 클래스(CountVectorizer)의 build_analyzer 메서드를 호출 - super로
    return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc)) # 만들어진 각 토큰에 대하여 lemmatization
```

```python
# 전체 텍스트를 리스트로 저장
text = list(train.text.values)

# vectorizer 불러오기
tf_vectorizer = LemmaCountVectorizer(max_df = 0.95, min_df=2, stop_words='english', decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

```python
feature_names = tf_vectorizer.get_feature_names_out() # term frequency vectorizer에서 단어 리스트 추출
count_vec = np.asarray(tf.sum(axis=0)).ravel() # 같은 문서 내 term freuquency를 합함, np array로 변환하고 1차원으로 압축
zipped = list(zip(feature_names, count_vec)) # 단어랑 빈도수를 pair로 저장
x, y = (list(x) for x in zip(*sorted(zipped, key = lambda x: x[1], reverse=True))) # unzip하고 내림차순 정렬

Y = np.concatenate([y[0:15], y[-16:-1]]) # top 15, bottom 15 frequent words
X = np.concatenate([x[0:15], x[-16:-1]])

data = [go.Bar(
    x = x[0:50],
    y = y[0:50],
    marker = dict(colorscale = 'Jet', color = y[0:50]),
    text = 'Word counts'
)]

layout = go.Layout(
    title = 'Top 50 Word frequencies after Preprocessing'
)
fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename = 'basic-bar')

data = [go.Bar(
    x = x[-100:],
    y = y[-100:],
    marker = dict(colorscale = 'Portland', color = y[-100:]),
    text = 'Word counts'
)]

layout = go.Layout(
    title = 'Bottom 100 Word frequencies after Preprocessing'
)

fig = go.Figure(data=data, layout = layout)

py.iplot(fig, filename='basic-bar')
```

### 3b. Latent Dirichlet Allocation

```python
lda = LatentDirichletAllocation(n_components = 11, max_iter = 5, learning_method = 'online', learning_offset = 50, random_state = 0)
```

```python
lda.fit(tf)
```

```python
n_top_words = 40
print("\nTopics in LDA model: ")
tf_feature_names = tf_vectorizer.get_feature_names_out()
print_top_words(lda, tf_feature_names, n_top_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```
