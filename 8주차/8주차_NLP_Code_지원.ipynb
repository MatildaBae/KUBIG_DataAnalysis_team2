## 3. Topic modelling
1. Latent Dirichlet Allocation : Assigning weights to words in corpus, each topic will assign different prob weights to each word
2. Non-negative Matrix Factorization : Take input matrix, approximates the factorization of this matrix into two other matrices, with the caveat that the values in the matrix be non-negative

```python
# Define helper function to print top words

def print_top_words(model, feature_names, n_top_words) :
    for index, topic in enumerate(model.components_) :
    # model.components_ : 주제 모델의 구성요소 행렬로, 각 주제에 대한 단어의 중요도 나타냄
    
        message = '\nTopic #{}'.format(index) # 현재 주제 번호에 메세지 추가
        
        message += ' '.join([feature_names[i] for i in topic.argsort()
                             [:-n_top_words - 1 : -1]])
        # 주제 벡터에서 값이 큰 순서대로 상위 'n_top_words'개의 단어 인덱스 정렬 추출
        
        print(message)
        print("="*70)
```

### Putting all the preprocessing steps together
### Extending the CountVectorizer class with a lemmatizer

```python
lemm = WordNetLemmatizer()

# CountVectorizer를 확장하여 텍스트 데이터를 벡터화할 때 단어의 기본 형태로 반환하는 클래스 정의
# 단어의 다양한 형태를 같은 단어로 취급
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        
        # build_analyzer 메소드 호출하여 기본 분석기 생성
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        
        # 분석된 단어들을 순회하며 각 단어를 기본 형태로 변환하는 람다 함수를 반환
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))   
```

```python
# Storing the entire training text in a list
text = list(train.text.values)

# Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df = 0.95, # 특정 단어가 95% 이상 문서에 나타나면 무시
                                   min_df = 2, # 최소 2개 이상 문서, 드물게 등장하는 단어 제거
                                   stop_words = 'english',
                                   decode_error = 'ignore' # 디코딩 오류 무시
                                    )

# 트레인 데이터 단어들의 기본 형태
tf = tf_vectorizer.fit_transform(text)
```

```python
# 벡터화된 텍스트 데이터의 단어 목록 추출
feature_names = tf_vectorizer.get_feature_names_out()

# 단어 빈도 계산
count_vec = np.asarray(tf.sum(axis=0)).ravel()

# 단어와 빈도 zip
zipped = list(zip(feature_names, count_vec))

# 단어와 빈도 정렬, 내림차순
x, y = (list(x) for x in zip(*sorted(zipped, key=lambda x: x[1], reverse=True)))

# Now I want to extract out on the top 15 and bottom 15 words
Y = np.concatenate([y[0:15], y[-16:-1]])
X = np.concatenate([x[0:15], x[-16:-1]])

# Plotting the Plot.ly plot for the Top 50 word frequencies
data = [go.Bar(
            x = x[0:50],
            y = y[0:50],
            marker= dict(colorscale='Jet',
                         color = y[0:50]),
            text='Word counts'
    )]

layout = go.Layout(
    title='Top 50 Word frequencies after Preprocessing')

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='basic-bar')

# Plotting the Plot.ly plot for the Top 50 word frequencies
data = [go.Bar(
            x = x[-100:],
            y = y[-100:],
            marker= dict(colorscale='Portland',
                         color = y[-100:]),
            text='Word counts'
    )]

layout = go.Layout(
    title='Bottom 100 Word frequencies after Preprocessing'
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='basic-bar')
```

### Latent Dirichlet Allocation

```python
# LDA : 텍스트 데이터를 주제별로 분류하는 과정
# 문서 집합에서 주제를 발견하고 각 문서에 대해 주제 분포를 추정하는 데 사용
lda = LatentDirichletAllocation(n_components=11, # 주제 11개
                                max_iter=5, # 최대 반복 횟수 5번
                                learning_method = 'online', # 온라인 학습방법
                                learning_offset = 50., # 학습 과정의 초기 정착 단계
                                random_state = 0)

# 벡터화된 텍스트 데이터를 LDA 모델에 입력하여 11개의 주제 추출하며
# 각 주제에 대해 상위 단어들을 출력하는 단어들을 얻을 수
lda.fit(tf)
```

### Topics generated by LDA

```python
n_top_words = 40
print("\nTopics in LDA model: ")
tf_feature_names = tf_vectorizer.get_feature_names_out()
print_top_words(lda, tf_feature_names, n_top_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```
